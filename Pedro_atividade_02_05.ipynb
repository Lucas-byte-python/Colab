

> Código "Cliente"




```python
class Cliente:
    def __init__(self, id_cliente, nome, idade, data_nascimento):
        self.id_cliente = id_cliente
        self.nome = nome
        self.idade = idade
        self.data_nascimento = data_nascimento
    def get_id_cliente(self, int):
        return self.id_cliente
    def set_id_cliente(self, id_cliente):
        self.id_cliente = id_cliente
    def get_nome(self, str):
        return self.nome
    def set_nome(self, nome):
        self.nome = nome
    def get_idade(self, int):
        return self.idade
    def set_idade(self, idade):
        self.idade = idade
    def get_data_nascimento(self,str):
        return self.data_nascimento
    def set_data_nascimento(self, data_nascimento):
        self.data_nascimento = data_nascimento
```

```python
class Livro:
  def __init__(self, codigo, id_livro, titulo, editora):
    self.codigo = codigo
    self.id_livro = id_livro
    self.titulo = titulo
    self.editora = editora
  def get_codigo(self):
    return self.codigo
  def set_codigo(self, codigo):
    self.codigo = codigo
  def get_id_livro(self):
    return self.id_livro
  def set_id_livro(self, id_livro):
    self.id_livro = id_livro
  def get_titulo(self):
    return self.titulo
  def set_titulo(self, titulo):
    self.titulo = titulo
  def get_editora(self):
    return self.editora
  def set_editora(self, editora):
    self.editora = editora
```
