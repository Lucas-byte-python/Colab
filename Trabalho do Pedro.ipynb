```python
# Questão 42
questão = "Questão 42"
print(questão)

a = float(input('Digite o primeiro segmento: '))
b = float(input('Digite o segundo segmento: '))
c = float(input('Digite o terceiro segmento: '))

if a < b+c and b < a+c and c < a+b:
    print('É um triangulo')

    if a == b == c: # ou a == b and b == c:
        print('Triângulo Equilatero')

    elif  a != b != c != a:  #ou a != b and b != c and c != a:
        print('Triângulo escaleno')

    elif a < 90 and b < 90 and c < 90:
        print('Triângulo acutângulo')

    elif a == b or b == c or a == c:
        print('Triângulo Isosceles')

else:
    print('Não é um triangulo')
```

```python
# Questão 43
questão = "Questão 43"
print(questão)

peso = float(input('Digite o seu peso: '))
altura = float(input('Digite a sua altura: '))
imc = peso / (altura ** 2) 
print('Com um peso de {} e uma altura de {} você tem um IMC de {:.2f}'.format(peso,altura,imc))

if imc < 18.5:
    print('Você está abaixo do peso!')

elif imc <= 25:
    print('Você está no peso ideal!')

elif imc <= 30:
    print('Você está com sobre peso!')

elif imc <= 40:
    print('Você esta com Obesidade!')

elif imc > 40:
    print('Procure um medico urgente!!! Você esta com Obesidade Morbida!')
```
