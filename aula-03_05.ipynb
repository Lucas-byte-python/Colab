

> **A classe Inseto**





```python
class Inseto:
  def __init__ (self ,id_inseto ,nome ,qtd_patas):
    self.id_inseto = id_inseto
    self.nome = nome
    self.qtd_patas = qtd_patas
  def get_id_inseto(self):
    return self.id_inseto
  def set_id_inseto(self, id_inseto):
    self.id_inseto = id_inseto
  def get_nome(self):
    return self.nome
  def set_nome(self, nome):
    self.nome = nome
  def get_qtd_patas(self):
    return self.qtd_patas
  def set_qtd_patas(self, qtd_patas):
    self.qtd_patas = qtd_patas
```



> **A classe Região**



```python
class Regiao:
  def __init__ (self, id,nome ,pais):
    self.id = id
    self.nome = nome
    self.pais = pais
  def get_id(self):
    return self.id
  def set_id(self, id):
    self.id = id
  def get_nome(self):
    return self.nome
  def set_nome(self, nome):
    self.nome = nome
  def get_pais(self):
    return self.pais
  def set_pais(self, pais):
    self.pais = pais
```
