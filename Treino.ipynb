**Classe Usuário**

```python
import mysql.connector
class usuario():
      def __init__(self, host = "localhost", user = "root", pwd = "", db = "banco_usuario"):	        
          self.host = host
          self.user = user
          self.pwd = pwd
          self.db = db 
      def conecta(self):
          self.con = mysql.connector.connect(host = self.host, user = self.user, password = self.pwd, database = self.db)
          self.cur = self.con.cursor()
      print("Conexão bem sucedida!")    
      def desconecta(self): 
          self.con.close()
      def executa_DQL(self, sql):	      
          self.conecta()
          self.cur.execute(sql)
          res = self.cur.fechall()
          self.desconecta()
          return res
      def executa_DML(self, sql):	
          self.conecta()
          self.cur.execute(sql)
          self.con.commit()
          self.desconecta()
```

**Classe Executora**

```python
class Executa():
  executa = usuario()
  executa.conecta()
  
```
