```python
def switch(arg, default_case, like=False):
    switcher ={
        "Monday": 1,
        "Tuesday": 2,
        "Wednesday": 3,
        "Thursday": 4,
        "Friday": 5,
        "Saturday": 6,
        "Sunday": 7
    }
    if like:
        try:
            return switcher[arg]
        except KeyError:
            for key in switcher.keys():
              if str(arg).lower() in str (key).lower():
                return switcher[key]
              else:
                return default_case
    else:
        return switcher.get(arg, default_case)
        
```
